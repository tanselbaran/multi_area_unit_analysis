```python
import numpy as np
from matplotlib import pyplot as plt
import mat73
from scipy import io
```

```python
data_dict = mat73.loadmat('/home/baran/Desktop/rTBY25/1_freely_behav_210827_101633/amplifier_res.mat')
```

```python
clusterNotes = data_dict['clusterNotes']
spikeTimes = data_dict['spikeTimes']
spikesByCluster = data_dict['spikesByCluster']
```

```python
singleUnits = [] 
for i, cluster in enumerate(clusterNotes):
    if cluster == ['single']:
        singleUnits.append(i)
```

```python
filepath = '/home/baran/Desktop/rTBY25/1_freely_behav_210827_101633/time.dat'
sample_rate = 20000
with open(filepath, 'rb') as fid:
    raw_array = np.fromfile(fid, np.int32)
time_file = raw_array / float(sample_rate) #converting from int32 to seconds
```

```python
spike_trains = np.zeros((len(singleUnits),len(time_file)))
```

```python
for i, unit in enumerate(singleUnits):
    clusterSpikes = np.asarray(spikesByCluster[unit]).astype('int32')[0]
    clusterSpikeTimes = spikeTimes[clusterSpikes-1]
    spike_trains[i][clusterSpikeTimes] = 1
```

```python
LFP_ripple = io.loadmat('/home/baran/Downloads/LFP_Ripple.mat')
lfp_data = LFP_ripple['lfp'][0][0][0]
lfp_timestamps = LFP_ripple['lfp'][0][0][1]
lfp_filtered = LFP_ripple['lfp'][0][0][2]
lfp_sample_rate = LFP_ripple['lfp'][0][0][3]
lfp_ripple = LFP_ripple['lfp'][0][0][4]
```

```python
sample_rate = 20000
pre = 0.020 #s
post = 0.020 #s
bin_size = 2 #ms
psth_range = np.arange(-pre*1000,post*1000,bin_size) #range for the interval of psth bins
bin_size_inds = int(sample_rate * bin_size/1000) #converting bin size from seconds to samples

evoked_psths = np.zeros((len(spike_trains), len(psth_range)))
evoked_trains = np.zeros((len(spike_trains),int((pre+post)*sample_rate)))

for unit in range(len(spike_trains)):
    for i, stim_ind in enumerate(lfp_ripple[1:,0]):
        evoked_trains[unit] = evoked_trains[unit] +  spike_trains[unit][(int(round((stim_ind-pre)*sample_rate))):(int(round((stim_ind+post)*sample_rate)))]
    evoked_trains[unit] = evoked_trains[unit] / (len(lfp_ripple)-1)
    for bin in range(len(psth_range)):
        evoked_psths[unit][bin] = np.sum(evoked_trains[unit][bin*bin_size_inds:(bin+1)*bin_size_inds])
```

```python
%matplotlib inline
for i in range(len(evoked_psths)):
    plt.figure()
    plt.title("Unit:"+str(singleUnits[i]))
    plt.xlabel('Time from trigger onset (ms)')
    plt.plot(psth_range,evoked_psths[i])
    plt.show()
```

```python
ripple_firing_rates = np.zeros((len(lfp_ripple),len(spike_trains)))
average_firing_rates = np.sum(spike_trains,axis=1)/(len(time_file)/20000)
for i in range(len(lfp_ripple)):
    ripple_begin = int(round(lfp_ripple[i,0]*20000))
    ripple_end = int(round(lfp_ripple[i,1]*20000))
    ripple_duration = lfp_ripple[i,1] - lfp_ripple[i,0]
    ripple_firing_rates[i] = np.sum(spike_trains[:,ripple_begin:ripple_end],axis=1)/ripple_duration
```

```python
for i in range(len(spike_trains)):
    plt.figure(figsize=(20,5))
    plt.title('Unit:'+str(singleUnits[i]))
    plt.scatter(np.arange(len(lfp_ripple)),ripple_firing_rates[:,i])
    plt.hlines(average_firing_rates[i],0,len(lfp_ripple),color='r')
    plt.xlabel('Ripple event')
    plt.ylabel('Firing rate spikes/s')
    plt.ylim([0,100])
    plt.show()
    print("Ratio of ripple events where firing rate is higher than average:"+str(len(np.where(ripple_firing_rates[:,i]>average_firing_rates[i])[0])/len(lfp_ripple)))
```
